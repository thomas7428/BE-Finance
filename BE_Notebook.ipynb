# Pricing des zéro-coupon dans un modèle à taux flottant : le modèle de Vasiček (Projet N° 1)

**Rapport de BE — Partie I : Le modèle et pricing des zéro-coupon**

Soit le taux court $r=(r(t))_{t\ge 0}$ donné comme unique solution de l'équation différentielle stochastique suivante :
$$
dr(t)=b(t,r(t))\,dt+\sigma(t,r(t))\,dW_t,\qquad t\ge0,\quad r(0)=r_0,
$$
où $b,\sigma$ sont des processus adaptés, $W$ est un mouvement brownien défini sur $(\Omega,\mathcal{F},\mathbb{P})$ et $r_0>0$. On note $(\mathcal{F}_t)_{t\ge0}$ la filtration naturelle de $W$ (satisfaisant les conditions usuelles).

Le compte money-market associé au taux $r$ est donné par :
$$
B(t)=\exp\!\Big(\int_0^t r(s)\,ds\Big),\qquad t\ge0.
$$

En supposant que $\mathbb{P}$ est une mesure martingale équivalente (MME) pour la numéraire $B$, le prix à la date $t$ d'un zéro-coupon de maturité $T$ s'exprime par :
$$
p(t,T)=\mathbb{E}\!\left[\left. e^{-\int_t^T r(s)\,ds} \right|\mathcal{F}_t\right],\qquad t\in[0,T],\quad p(T,T)=1.
$$

Dans la suite, nous préciserons les hypothèses conduisant au modèle de Vasiček, résoudrons l'EDS du taux court et calculerons la formule fermée (ou semi-fermée) du prix $p(t,T)$.

**Question 1 :**

On fixe T>0 et on pose
$$
M_t:=p(t,T)B(t)^{-1}.
$$


a) Montrons que $M(t), t∈[0,T]$ est une martingale.

Par définition du prix du zéro‑coupon,
$$
p(t,T)=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
Donc car $B(t)=e^{\int_0^t r(s)\,ds}$ est $\mathcal{F}_t$-mesurable,
$$
M_t=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big]\;e^{-\int_0^t r(s)\,ds}
=\mathbb{E}\!\Big[\Big.e^{-\int_0^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
En notant $M_T=e^{-\int_0^T r(s)\,ds}$, on a donc $M_t=\mathbb{E}[M_T\mid\mathcal{F}_t]$. Par la "Tower Property" de l'espérance conditionnelle : 
$$
\mathbb{E}[M_u\mid\mathcal{F}_t]=\mathbb{E}\big[\mathbb{E}[M_T\mid\mathcal{F}_u]\big|\mathcal{F}_t\big]=\mathbb{E}[M_T\mid\mathcal{F}_t]=M_t.
$$
Ainsi $M$ est une martingale (sous l'hypothèse d'intégrabilité suffisante de $M_T$).

b) On a que $M=(M_t)_{t\in[0,T]}$ est une martingale adaptée à la filtration naturelle de $W$ et $M_T\in L^2$, 



On a de plus le théorème de représentation des martingales :
___
Soit $B_t$ un mouvement brownien sur un espace de probabilité filtré standard $(\Omega,\mathcal{F},(\mathcal{G}_t)_{t\ge0},\mathbb{P})$, et soit $\mathcal{G}_t$ la filtration augmentée engendrée par $B$. Si $X\in L^2(\Omega,\mathcal{G}_\infty,\mathbb{P})$, alors il existe un processus prévisible $C$ adapté à $(\mathcal{G}_t)$ tel que
$$
X=\mathbb{E}[X]+\int_0^\infty C_s\,dB_s.
$$

Par conséquent,
$$
\mathbb{E}[X\mid\mathcal{G}_t]=\mathbb{E}[X]+\int_0^t C_s\,dB_s.
$$
___

En appliquant ce résultat à $X=M_T$ et en prenant $B=W$, il existe donc un processus prévisible $v=(v(t))_{t\in[0,T]}$ tel que
$$
M_t=M_0+\int_0^t v(s)\,dW_s,\qquad t\in[0,T],
$$
avec $\int_0^T v(s)^2\,ds<\infty$ p.s., et $v$ est unique à modifications négligeables.

**Question 2:**

Partant de $M_t=\dfrac{p(t,T)}{B(t)}$, on a $p(t,T)=M_tB(t)$. Par la formule d'Ito pour le produit de semimartingales :
$$
dp(t,T)=B(t)\,dM_t+M_t\,dB(t)+d\langle M,B\rangle_t.
$$
Or $B(t)=\exp\!\big(\int_0^t r(s)\,ds\big)$ est de variation finie, donc $dB(t)=B(t)r(t)\,dt$ ce qui nous donne aussi $\langle M,B\rangle_t=0$. 

De plus, d'après la représentation martingale $dM_t=v(t)\,dW_t$. On obtient donc :
$$
dp(t,T)=B(t)v(t)\,dW_t+M_tB(t)r(t)\,dt
= v(t)B(t)\,dW_t + p(t,T)r(t)\,dt,
$$
pour $t\in[0,T]$, ce qui donne bien l'expression souhaitée.

**Question 4 :**

On suppose qu'il existe une fonction $F_T:[0,T]\times\mathbb{R}\to\mathbb{R}$ telle que $p(t,T)=F_T(t,r(t))$. On impose la régularité classique pour pouvoir appliquer la formule d'Itô et faire des égalités terme à terme :
- $F_T\in C^{1,2}([0,T)\times\mathbb{R})$ (continue en $t$, deux fois dérivable en $r$),
- croissances/conditions d'intégrabilité suffisantes (par exemple croissance au plus polynomiale) pour garantir l'application d'Itô et l'appartenance en $L^2$ de $M_T$.

Dérivation : par Itô appliqué à $F_T(t,r(t))$ avec $dr(t)=b(t,r(t))dt+\sigma(t,r(t))dW_t$,
$$
dF_T(t,r(t))=\Big(\partial_t F_T+b\,\partial_r F_T+\tfrac12\sigma^2\partial_r^2 F_T\Big)\,dt
+\sigma\,\partial_r F_T\,dW_t.
$$
Or, comme $p(t,T)=F_T(t,r(t))$ et d'après la question 2 on a aussi
$$
dp(t,T)=p(t,T)r(t)\,dt+v(t)B(t)\,dW_t.
$$
En identifiant les termes en $dt$ et $dW_t$ on obtient l'EDP et l'expression de la composante stochastique :
EDP (pour tout $(t,r)\in[0,T)\times\mathbb{R}$)

Identification des termes en $dW_t$ :
$$
\boxed{\sigma(t,r)\,\partial_r F_T(t,r)=v(t)\,B(t)}
$$
Et pour $dt$ : 
$$
\ \partial_t F_T(t,r)+b(t,r)\,\partial_r F_T(t,r)+\tfrac12\sigma(t,r)^2\,\partial_r^2 F_T(t,r)-r\,F_T(t,r)=0\
$$
condition terminale (au bord) :
$$
\boxed{F_T(T,r)=1\quad\text{pour tout }r\in\mathbb{R}.}
$$

Nature de l'EDP : il s'agit d'une EDP linéaire du second ordre, parabolique (rétrograde / « backward » en temps) — problème de Cauchy terminal.
Preuve que M est une martingale (suffisance) : posons
$$
M_t:=\frac{F_T(t,r(t))}{B(t)}.
$$
Comme $dB(t)=B(t)r(t)\,dt$ et $B$ est de variation finie donc $\langle F_T(\cdot,r(\cdot)),B\rangle\equiv0$, on obtient
$$
dM_t=\frac{1}{B(t)}\Big(dF_T(t,r(t))-F_T(t,r(t))\,r(t)\,dt\Big).
$$
Appliquant la formule d'Itô à $F_T(t,r(t))$ :
$$
dF_T=\big(\partial_t F_T+b\,\partial_r F_T+\tfrac12\sigma^2\partial_r^2 F_T\big)\,dt
+\sigma\,\partial_r F_T\,dW_t.
$$
Par l'EDP $\partial_t F_T+b\partial_r F_T+\tfrac12\sigma^2\partial_r^2 F_T - rF_T=0$, le terme en $dt$ de $dF_T-F_Tr\,dt$ s'annule. Il reste donc
$$
dM_t=\frac{\sigma(t,r(t))\,\partial_r F_T(t,r(t))}{B(t)}\,dW_t=:v(t)\,dW_t.
$$
Ainsi $M$ est une intégrale stochastique sans dérive. Sous les hypothèses de régularité et de croissance (assurant $\mathbb{E}[\int_0^T v(t)^2dt]<\infty$), $M$ est une martingale $L^2$. On a explicitement
$$
v(t)=\frac{\sigma(t,r(t))\,\partial_r F_T(t,r(t))}{B(t)}.
$$

Remontée (sens inverse de la question 1) — justification précise à partir de l'EDP sur le terme en $dW_t$ :
- Partir de l'EDP et de l'identification $\sigma(t,r)\partial_r F_T(t,r)=v(t)B(t)$. Définir alors $M_t:=F_T(t,r(t))/B(t)$.
- Appliquer Itô à $F_T(t,r(t))$ et utiliser $dB(t)=B(t)r(t)\,dt$ pour obtenir l'expression de $dM_t$. Par l'EDP, le terme en $dt$ disparaît et on obtient exactement $dM_t=v(t)\,dW_t$.
- Sous les hypothèses de croissance/reguralité, $\mathbb{E}\big[\int_0^T v(t)^2\,dt\big]<\infty$ (car $\sigma\partial_r F_T$ a croissance contrôlée et $B(t)^{-1}=e^{-\int_0^t r(s)ds}$ est adapté), donc $M$ est une martingale L^2 (stochastique intégrale carrée-intégrable).
- La condition terminale $F_T(T,r)=1$ donne $M_T=B(T)^{-1}=e^{-\int_0^T r(s)\,ds}$, dont l'intégrabilité est garantie par les hypothèses ; ainsi la martingale est bien définie sur $[0,T]$.

Conclusion : en remontant depuis l'identification du terme en $dW_t$ (l'EDP donnée), l'application d'Itô fournit $dM_t=v(t)\,dW_t$ avec $v\in L^2$, donc $M$ est une martingale, ce qui est l'inverse de l'argument utilisé en question 1.
