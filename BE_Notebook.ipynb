# Pricing des zéro-coupon dans un modèle à taux flottant : le modèle de Vasiček (Projet N° 1)

## Partie I : Le modèle et pricing des zéro-coupon

Soit le taux court $r=(r(t))_{t\ge 0}$ donné comme unique solution de l'équation différentielle stochastique suivante :
$$
dr(t)=b(t,r(t))\,dt+\sigma(t,r(t))\,dW_t,\qquad t\ge0,\quad r(0)=r_0,
$$
où $b,\sigma$ sont des processus adaptés, $W$ est un mouvement brownien défini sur $(\Omega,\mathcal{F},\mathbb{P})$ et $r_0>0$. On note $(\mathcal{F}_t)_{t\ge0}$ la filtration naturelle de $W$ (satisfaisant les conditions usuelles).

Le compte money-market associé au taux $r$ est donné par :
$$
B(t)=\exp\!\Big(\int_0^t r(s)\,ds\Big),\qquad t\ge0.
$$

En supposant que $\mathbb{P}$ est une mesure martingale équivalente (MME) pour la numéraire $B$, le prix à la date $t$ d'un zéro-coupon de maturité $T$ s'exprime par :
$$
p(t,T)=\mathbb{E}\!\left[\left. e^{-\int_t^T r(s)\,ds} \right|\mathcal{F}_t\right],\qquad t\in[0,T],\quad p(T,T)=1.
$$

Dans la suite, nous préciserons les hypothèses conduisant au modèle de Vasiček, résoudrons l'EDS du taux court et calculerons la formule fermée (ou semi-fermée) du prix $p(t,T)$.

**Question 1 :**

On fixe T>0 et on pose
$$
M_t:=p(t,T)B(t)^{-1}.
$$


a) Montrons que $M(t), t∈[0,T]$ est une martingale.

Par définition du prix du zéro‑coupon,
$$
p(t,T)=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
Donc car $B(t)=e^{\int_0^t r(s)\,ds}$ est $\mathcal{F}_t$-mesurable,
$$
M_t=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big]\;e^{-\int_0^t r(s)\,ds}
=\mathbb{E}\!\Big[\Big.e^{-\int_0^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
En notant $M_T=e^{-\int_0^T r(s)\,ds}$, on a donc $M_t=\mathbb{E}[M_T\mid\mathcal{F}_t]$. Par la "Tower Property" de l'espérance conditionnelle : 
$$
\mathbb{E}[M_u\mid\mathcal{F}_t]=\mathbb{E}\big[\mathbb{E}[M_T\mid\mathcal{F}_u]\big|\mathcal{F}_t\big]=\mathbb{E}[M_T\mid\mathcal{F}_t]=M_t.
$$
Ainsi $M$ est une martingale (sous l'hypothèse d'intégrabilité suffisante de $M_T$).

b) On a que $M=(M_t)_{t\in[0,T]}$ est une martingale adaptée à la filtration naturelle de $W$ et $M_T\in L^2$, 



On a de plus le théorème de représentation des martingales :
___
Soit $B_t$ un mouvement brownien sur un espace de probabilité filtré standard $(\Omega,\mathcal{F},(\mathcal{G}_t)_{t\ge0},\mathbb{P})$, et soit $\mathcal{G}_t$ la filtration augmentée engendrée par $B$. Si $X\in L^2(\Omega,\mathcal{G}_\infty,\mathbb{P})$, alors il existe un processus prévisible $C$ adapté à $(\mathcal{G}_t)$ tel que
$$
X=\mathbb{E}[X]+\int_0^\infty C_s\,dB_s.
$$

Par conséquent,
$$
\mathbb{E}[X\mid\mathcal{G}_t]=\mathbb{E}[X]+\int_0^t C_s\,dB_s.
$$
___

En appliquant ce résultat à $X=M_T$ et en prenant $B=W$, il existe donc un processus prévisible $v=(v(t))_{t\in[0,T]}$ tel que
$$
M_t=M_0+\int_0^t v(s)\,dW_s,\qquad t\in[0,T],
$$
avec $\int_0^T v(s)^2\,ds<\infty$ p.s., et $v$ est unique à modifications négligeables.

**Question 2:**

Partant de $M_t=\dfrac{p(t,T)}{B(t)}$, on a $p(t,T)=M_tB(t)$. Par la formule d'Ito pour le produit de semimartingales :
$$
dp(t,T)=B(t)\,dM_t+M_t\,dB(t)+d\langle M,B\rangle_t.
$$
Or $B(t)=\exp\!\big(\int_0^t r(s)\,ds\big)$ est de variation finie, donc $dB(t)=B(t)r(t)\,dt$ ce qui nous donne aussi $\langle M,B\rangle_t=0$. 

De plus, d'après la représentation martingale $dM_t=v(t)\,dW_t$. On obtient donc :
$$
dp(t,T)=v(t)B(t)\,dW_t+M_tB(t)r(t)\,dt \\
\boxed{dp(t,T)= v(t)B(t)\,dW_t + p(t,T)r(t)\,dt}
$$
pour $t\in[0,T]$, ce qui donne bien l'expression souhaitée.

$$
dM_t = e^{-\int_0^t r(s)\,ds}\cdot dF_T + F_T\cdot d(e^{-\int_0^t r(s)\,ds}) + \langle dF_T , d(e^{-\int_0^t r(s)\,ds}) \rangle

De plus on a : $dg(t) = r(t) \cdot dt + 0 \cdot dW_t$
Nous n'avons donc pas de variations en $dW_t$ ce qui implique que nous n'avons pas de terme d'erreur quadratique.

Avec la formule d'Itô :
$$
dF_T = \partial_t F_T \cdot dt + \partial_r F_T \cdot dr + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 \cdot dt

De plus : $d(e^{-g(t)}) = -r(t) \cdot e^{-g(t)} + 0 \cdot dW_t$ d'où $<...,...> = 0$

Ainsi : 
$$
dM_t = e^{-\int_0^t r(s)\,ds}\cdot (\partial_t F_T + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 - r(t) \cdot F_T)dt + e^{-\int_0^t r(s)\,ds} \cdot \partial_r F_T \cdot dr
$$
et $dr = b \cdot dt + \sigma \cdot dW_t$

D'où :
$$
\begin{aligned}
dM_t = \phi_t \cdot dW_t &\Leftrightarrow e^{-\int_0^t r(s)\,ds} \cdot (\partial_t F_T + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 - r(t) \cdot F_T + \partial_r F_T \cdot b) = 0 \\
&\Leftrightarrow \partial_t F_T + b \cdot \partial_r F_T + 1/2 \cdot \sigma^2 \cdot \partial_r^2 F_T + r \cdot F_T = 0
\end{aligned}
$$
De plus on a la condition terminale $F_T(T,r) = 1$ pour cette équation parabolique.
D'où pour résumer :
$$
\boxed{
\begin{cases}
\partial_t F_T + b \cdot \partial_r F_T + 1/2 \cdot \sigma^2 \cdot \partial_r^2 F_T + r \cdot F_T = 0 \\
F_T(T,r) = 1
\end{cases}
}
$$

## Partie II : Le modèle de Vasiček

**Question 1 :**

On a : $dr(t) = (b + \beta \cdot r(t))dt + \sigma \cdot dW_t $
$$
Y_t := e^{-\beta t} r(t)
$$
Par Itô :
$$
\begin{aligned}
dY_t &= - \beta e^{-\beta t} r(t) \cdot dt + e^{-\beta t} dr(t) + 0 \\
dY_t &= - \beta e^{-\beta t} r(t) \cdot dt + e^{-\beta t} b \cdot dt + e^{-\beta t} \beta r(t) \cdot dt + e^{-\beta t} \sigma dW_t \\
Y_t &= Y_0 + b \int_0^t e^{-\beta s}\,ds + \sigma \int_0^t e^{-\beta s}\,dW_s
\end{aligned}
$$
En inversant : $r_t = Y_r e^{\beta t}$
D'où :
$$
r_t = Y_0 e^{\beta t} + b e^{\beta t} \int_0^t e^{-\beta s}\,ds + \sigma e^{\beta t} \int_0^t e^{-\beta s}\,dW_s \\
r_t = \underbrace {r_0 e^{\beta t} + b e^{\beta t} \int_0^t e^{-\beta s}\,ds }_\textrm{Partie déterministe} + \underbrace {\sigma e^{\beta t} \int_0^t e^{-\beta s}\,dW_s}_\textrm{Partie gaussienne}
$$
En supposant $\beta$ non nul (le cas $\beta = 0$ correspond à la partie I): 
$$
b e^{\beta t} \int_0^t e^{-\beta s}\,ds = b e^{\beta t} \left[ \frac{e^{-\beta s}}{-\beta} \right]^t_0 = b e^{\beta t} \cdot \frac{e^{-\beta t} - 1}{-\beta}
$$

D'où :
$$
\boxed{r_t = \underbrace {r_0 e^{\beta t} + b \cdot \frac{e^{\beta t} - 1}{\beta}}_\textrm{Partie déterministe} + \underbrace {\sigma e^{\beta t} \int_0^t e^{-\beta s}\,dW_s}_\textrm{Partie aléatoire}}
$$

La partie aléatoire de $r_t$ suit donc une loi normale centrée. Puis la partie déterministe n'influence que l'espérance.

Donc $r_t \sim N(r_0 e^{\beta t} + b \cdot \frac{e^{\beta t} - 1}{\beta}, \sigma ^2\int_0^t e^{2\beta (t-s)}ds)$.

Ce modèle n'est pas représentatif de la réalité car il permet à $r_t$ d'être négatif. Ce n'est pas réaliste. On pourrait penser à utiliser par la suite des modèles avec $r_t$ suivant une loi de Weibull.

**Question 4 :**

On résout $B$ puis $A$ en remplaçant $F$ dans l'EDP.

On obtient pour $B$ l'équation $(E)$: $\partial_t B + \beta B = 0$

D'où la solution homogène : $B_H (t) = K  \cdot e^{-\beta t}$

On utilise la méthode de variation de la constante : 
$$
\begin{aligned}
B(t,T) &= \lambda(t) \cdot e^{- \beta t} \\
B'(t,T) &= \lambda'(t) \cdot e^{- \beta t} -\beta \lambda(t) \cdot e^{- \beta t}
\end{aligned}
$$

D'après l'équation $(E)$ on a :
$$
\begin{aligned}
\lambda'(t) \cdot e^{- \beta t} &= -1 \\
\lambda'(t) &= - \cdot e^{\beta t} \\
\lambda(t) &= - \int^t_0 \cdot e^{\beta s} ds \\
\lambda(t) &= - \left[ \frac{e^{\beta s}}{\beta} \right]^t_0 = \frac{1 - e^{\beta t}}{\beta}
\end{aligned}
$$
On obtient donc finalement : 
$$
B(t,T) = K \cdot e^{-\beta t} + \frac{e^{-\beta t} - 1}{\beta}
$$
Avec les conditions aux limites on a :
$$ 
\begin{aligned}
B(T,T) &= 0 \\
K \cdot e^{-\beta T} + \frac{e^{-\beta T} - 1}{\beta} &= 0 \\
K &= \frac{1 - e^{-\beta T}}{\beta} \cdot e^{\beta T} \\
K &=  \frac{e^{\beta T} - 1}{\beta}
\end{aligned}
$$

D'où finalement : 
$$
\begin{aligned}
B(t,T) &= \frac{e^{\beta T} - 1}{\beta} \cdot e^{- \beta t} + \frac{e^{- \beta t} - 1}{\beta} \\
&= e^{- \beta t} \left[\frac{e^{\beta T} - 1}{\beta} + \frac{1}{\beta} \right] - \frac{1}{\beta}
\end{aligned}
$$
D'où : $$ \boxed{B(t,T) = \frac{e^{\beta \left(T-t \right)} - 1}{\beta}}$$

Pour A on procède de même en reprenant l'EDP :
$$
\begin{aligned}
\partial_t A &= - \frac{\sigma^2}{2} B^2 + b \cdot B \\
\partial_t A &= - \frac{\sigma^2}{2} \frac{\left(e^{\beta \left(T-t \right)} - 1 \right)^2}{\beta^2}  + \frac{b}{\beta} \left(e^{\beta \left(T-t \right)} - 1 \right) \\

\partial_t A &= - \frac{\sigma^2}{2 \beta^2} e^{2 \beta \left(T-t \right)} + \left( \frac{b}{\beta} + \frac{\sigma^2}{\beta^2} \right) e^{\beta \left(T - t \right)} - \left( \frac{\sigma^2}{2 \beta^2} + \frac{b}{\beta} \right) \\
\end{aligned}
$$
On obtient alors après intégration : 
$$
\begin{aligned}
A(T)-A(t) &= - \frac{\sigma^2}{2 \beta^2} \left[ \frac{e^{2 \beta \left(T-t \right)}}{-2 \beta} \right]^T_t + \left( \frac{b}{\beta} + \frac{\sigma^2}{\beta^2} \right) \left[ \frac{e^{\beta \left(T - t \right)}}{- \beta}\right]^T_t - \left( \frac{\sigma^2}{2 \beta^2} + \frac{b}{\beta} \right) \left[t\right]^T_t \\
A(T)-A(t) &= - \frac{\sigma^2}{2 \beta^2} \left( \frac{e^{2 \beta \left(T-t \right)} - 1}{2 \beta} \right) + \left( \frac{b}{\beta} + \frac{\sigma^2}{\beta^2} \right) \left (\frac{e^{\beta \left(T - t \right)} - 1}{ \beta}\right) - \left( \frac{\sigma^2}{2 \beta^2} + \frac{b}{\beta} \right) \left(T-t\right) \\
A(t) &= \frac{\sigma^2}{4 \beta^3} \left(e^{2 \beta \left(T-t \right)} - 1 \right) + \left(\frac{\beta b + \sigma^2}{\beta^3} \right) \left (e^{\beta \left(T - t \right)} - 1\right) - \left( \frac{2\beta b + \sigma^2}{2 \beta^2} \right) \left(T-t\right) \\
\end{aligned}
$$

On obtient alors la solution explicite suivante :
$$
\boxed{F_T(t,r) = e^{A(t,T) - B(t,T)r}} \\
\textrm{avec} \boxed{ \begin{cases}

B(t,T) = \frac{e^{\beta \left(T-t \right)} - 1}{\beta}

}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

#  Paramètres de la simulation
b = 0.02        
sigma = 0.03    
beta = -0.5
r0 = 0.01       
T =4         
dt = 1/252      
N_steps = int(T / dt)
t_grid = np.linspace(0, T, N_steps+1)
n_paths = 1000  

def B(t, T):
    return (np.exp(beta*(T-t))-1)/beta

def A(t,T):
    T1 = (sigma**2/(4*beta**3))*(np.exp(2*beta*(T-t))-1)
    T2 = ((b/beta**2 + (sigma**2)/(beta**3)))*(np.exp(beta*(T-t))-1)
    T3 = ((sigma**2 / (2*beta**2))+b/beta)*(T-t)
    return T1-T2+T3

def simulate_paths_euler(b, sigma, r0, T, dt, n_paths, theta_shift=None): # Théta shift pour IS de la partie III
    N = int(np.round(T/dt))
    t = np.linspace(0, T, N+1)
    r = np.zeros((n_paths, N+1))
    r[:,0] = r0
    sqrt_dt = np.sqrt(dt)
    W_T_sim = np.zeros(n_paths)
    
    for k in range(N):
        # dW ~ sqrt(dt) * N(0,1)
        Z = np.random.randn(n_paths)
        if theta_shift is not None:
            Z = Z + theta_shift * sqrt_dt  # translation des incréments (IS)
        # mise à jour du W discret
        W_T_sim += Z * sqrt_dt
        drift = b + beta * r[:, k]
        r[:, k + 1] = r[:, k] + drift * dt + sigma * sqrt_dt * Z
    return t, r, W_T_sim

t, r_paths, W_T_sim = simulate_paths_euler(b, sigma, r0, T, dt, n_paths)

B_grid = B(t,T)
A_grid = A(t,T)

p_paths = np.exp( A_grid[np.newaxis, :] - (B_grid[np.newaxis, :] * r_paths) )

n_plot = 5
fig, ax = plt.subplots(2,1, figsize=(10,8), sharex=True)

for i in range(min(n_plot, n_paths)):
    ax[0].plot(t, r_paths[i,:], lw=0.8, alpha=0.7)
ax[0].set_title('Trajectoires simulées de r(t) (Euler-Maruyama)')
ax[0].set_ylabel('r(t)')

for i in range(min(n_plot, n_paths)):
    ax[1].plot(t, p_paths[i,:], lw=0.8, alpha=0.7)
ax[1].set_title(f'Prix du zéro-coupon p(t;T={T}) pour chaque trajectoire')
ax[1].set_ylabel('p(t;T)')
ax[1].set_xlabel('t (années)')

plt.tight_layout()
plt.show()


```

## Partie III : Estimation de Monte-Carlo dans le modèle de Vasiček

**Question 1 :**

Le contrat considéré est une **option d'achat (call)** sur le taux court $r_T$ avec un prix d'exercice $k$. Son payoff à maturité $T$ est :
$$
\Phi(r_T) = (r_T - k)^+
$$

Sous l'hypothèse que $\mathbb{P}$ est une mesure martingale équivalente (MME) pour le numéraire $B(t) = e^{\int_0^t r(s)\,ds}$, le prix du contrat à la date $t \in [0, T]$ est donné par l'espérance conditionnelle du payoff actualisé :

$$
\boxed{C_t = \mathbb{E}\left[\left. e^{-\int_t^T r(s)\,ds} (r_T - k)^+ \right| \mathcal{F}_t\right], \quad t \in [0, T]}
$$

avec la condition terminale $C_T = (r_T - k)^+$.

**Question 2 :** 

De manière similaire à la Partie I, on suppose qu'il existe une fonction $G: [0,T] \times \mathbb{R} \to \mathbb{R}$ telle que :
$$
C_t = G(t, r(t)), \quad t \in [0, T]
$$

Démarche :

1. **Processus actualisé :** On considère le processus actualisé
$$
\tilde{C}_t := C_t B(t)^{-1} = G(t, r(t)) e^{-\int_0^t r(s)\,ds}
$$

2. **Propriété de martingale :** Sous la MME $\mathbb{P}$, le processus $\tilde{C}_t$ est une martingale. En effet :
$$
\tilde{C}_t = \mathbb{E}\left[\left. e^{-\int_0^T r(s)\,ds} (r_T - k)^+ \right| \mathcal{F}_t\right]
$$

3. **Application de la formule d'Itô :** En appliquant la formule d'Itô au produit $\tilde{C}_t = G(t, r(t)) \cdot e^{-\int_0^t r(s)\,ds}$, on obtient :
$$
d\tilde{C}_t = e^{-\int_0^t r(s)\,ds} \left[\frac{\partial G}{\partial t} + (b + \beta r) \frac{\partial G}{\partial r} + \frac{\sigma^2}{2} \frac{\partial^2 G}{\partial r^2} - r \cdot G\right] dt + e^{-\int_0^t r(s)\,ds} \sigma \frac{\partial G}{\partial r} dW_t
$$

4. **Condition de martingale :** Pour que $\tilde{C}_t$ soit une martingale, le terme en $dt$ (drift) doit être nul. Cela conduit à l'**EDP parabolique** suivante :

$$
\boxed{\frac{\partial G}{\partial t} + (b + \beta r) \frac{\partial G}{\partial r} + \frac{\sigma^2}{2} \frac{\partial^2 G}{\partial r^2} - r \cdot G = 0, \quad (t, r) \in [0, T) \times \mathbb{R}}
$$

avec la **condition terminale** :
$$
\boxed{G(T, r) = (r - k)^+, \quad r \in \mathbb{R}}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Paramètres (on reprend les fonctions A, B et simulate_paths_euler déjà définies plus haut)
# b, sigma, beta, r0, T, dt, n_paths doivent déjà exister dans le notebook (sinon redéfinir ici)
# ...existing code...

# --- Paramètres du contrat (modifier si besoin) ---
k = 0.04   # strike de l'option (exemple)
# On utilise t, r_paths calculés par simulate_paths_euler
np.random.seed(42)

# Simuler (vecteur de trajectoires)
t, r_paths, W_T_sim = simulate_paths_euler(b, sigma, r0, T, dt, n_paths)  # utilise la fonction existante
# t shape (N+1,), r_paths shape (n_paths, N+1)

# Intégrale du taux sur [0,T] pour chaque trajectoire (trapezoïde pour précision)
integral_r = np.trapezoid(r_paths, t, axis=1)        # shape (n_paths,)
r_T = r_paths[:, -1]                             # taux à maturité pour chaque trajectoire

# Payoff actualisé
payoffs = np.maximum(r_T - k, 0.0)
discounted_payoffs = np.exp(-integral_r) * payoffs

# Estimation Monte-Carlo
C0_MC = discounted_payoffs.mean()
std_error = discounted_payoffs.std(ddof=1) / np.sqrt(len(discounted_payoffs))
conf_interval = 1.96 * std_error

print("="*60)
print("PRIX MONTE-CARLO (call sur r_T)")
print("="*60)
print(f"C0_MC = {C0_MC:.8e}")
print(f"SE    = {std_error:.8e}")
print(f"IC95  = [{C0_MC - conf_interval:.8e}, {C0_MC + conf_interval:.8e}]")
if C0_MC > 0:
    print(f"Erreur relative = {100 * std_error / C0_MC:.4f}%")
else:
    print("Erreur relative = N/A (estimateur nul)")

# --- Diagnostics et graphiques ---
fig, ax = plt.subplots(2,1, figsize=(10,8))

# 1) Histogramme des payoffs actualisés
ax[0].hist(discounted_payoffs, bins=80, density=True, alpha=0.7, edgecolor='k')
ax[0].axvline(C0_MC, color='r', linestyle='--', label=f'Moyenne={C0_MC:.4e}')
ax[0].set_title('Distribution des payoffs actualisés')
ax[0].legend()
ax[0].grid(alpha=0.3)

# 2) Convergence cumulée
cumulative_mean = np.cumsum(discounted_payoffs) / np.arange(1, len(discounted_payoffs) + 1)
ax[1].plot(cumulative_mean, lw=1)
ax[1].axhline(C0_MC, color='r', ls='--')
ax[1].set_title("Convergence de l'estimateur Monte‑Carlo")
ax[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Affichage rapide des statistiques
print("\nStatistiques rapides des payoffs actualisés :")
print(f"  min = {discounted_payoffs.min():.8e}, max = {discounted_payoffs.max():.8e}")
print(f"  médiane = {np.median(discounted_payoffs):.8e}, std = {discounted_payoffs.std(ddof=1):.8e}")
print(f"  proportion in-the-money = {100 * (payoffs > 0).mean():.2f}%")
```

Pour trouver un cas où Monte-Carlo simple ne suffit pas, prenons un cas rare.
On applique donc des paramètres pour un cas rare.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Paramètres du modèle ---
b = 0.02
sigma = 0.015
beta = -0.5
r0 = 0.01
T = 4
dt = 1 / 252
k = 0.08  # Strike de l'option
# Pour les itérations sur N
N = 500                 # Première valeur de N (les suivantes seront de la forme N*N_Factor^i)
N_Factor = 1.5          # Valeur de l'expansion géométrique
Exponential_Max = 35    # Nombre maximal d'itérations
# Pour détecter la convergence
se_target = 2*1e-7    # Seuil d'erreur standard pour la convergence
diff_target = 1e-10  # Seuil de différence de moyenne pour la convergence
counter_threshold = 3   # Nombre d'itérations consécutives sous le seuil pour valider la convergence


def run_batch(n_paths, seed, theta_shift=None):
    """
    Effectue une simulation Monte-Carlo pour un nombre donné de trajectoires.
    """
    np.random.seed(seed)
    # now receive W_T simulated from simulate_paths_euler
    t, r_paths, W_T_sim = simulate_paths_euler(b, sigma, r0, T, dt, n_paths, theta_shift=theta_shift)
    integral_r = np.trapezoid(r_paths, t, axis=1)
    r_T = r_paths[:, -1]

    # Payoff actualisé
    payoffs = np.maximum(r_T - k, 0.0)
    discount_factors = np.exp(-integral_r)

    if theta_shift is None:
        # Monte-Carlo standard
        discounted_payoffs = discount_factors * payoffs
        weights = np.ones_like(payoffs)
    else:
        # Importance Sampling — poids cohérents avec la translation Z -> Z + theta*sqrt(dt)
        exponent = -theta_shift * W_T_sim + 0.5 * (theta_shift**2) * T
        weights = np.exp(np.clip(exponent, -700, 700))
        discounted_payoffs = weights * discount_factors * payoffs

    # Calcul des statistiques
    stats = {
        "mean": np.mean(discounted_payoffs),
        "std": np.std(discounted_payoffs, ddof=1),
        "se": np.std(discounted_payoffs, ddof=1) / np.sqrt(n_paths),
        "ess": (np.sum(weights)**2) / np.sum(weights**2) if theta_shift is not None else None,
    }
    return stats

# --- Grille de simulations ---
Ns = [int(N_Factor**i * N) for i in range(Exponential_Max)]
seed = 42
theta = (k - (np.exp(beta * T) * r0 + (b / beta) * (np.exp(beta * T) - 1))) / (sigma * np.sqrt(T))
# On va simuler jusqu'à ce que l'erreur standard soit suffisamment petite
N_mc_converge, N_is_converge = None, None
results_is, results_mc = [], []
counter_is, counter_mc = 0, 0

# --- Affichage des résultats ---
print("\n--- Résumé des résultats ---")
print(f"{'N':>7} | {'MC mean':>12} {'SE':>10} || {'IS mean':>12} {'SE':>10} {'ESS - IS':>10}")
print("-" * 70)
for i in range(len(Ns)):
    ns = Ns[i]
    # Arrêt si les deux méthodes ont convergé
    if N_mc_converge is None or N_is_converge is None:
        results_mc.append(run_batch(ns, seed, theta_shift=None))
        results_is.append(run_batch(ns, seed, theta_shift=theta))
    else:
        break
    
    # Vérification de la convergence
    if results_mc[-1]["se"] <= se_target or (i>0 and abs(results_mc[-1]['mean'] - results_mc[-2]['mean']) < diff_target):
        counter_mc += 1
        if counter_mc >= counter_threshold and N_mc_converge is None:
            N_mc_converge = ns
    if results_is[-1]["se"] <= se_target or (i>0 and abs(results_is[-1]['mean'] - results_is[-2]['mean']) < diff_target):
        counter_is += 1
        if counter_is >= counter_threshold and N_is_converge is None:
            N_is_converge = ns
    
    # Affichage de la ligne de résultats
    ess = results_is[-1]["ess"] if results_is[-1].get("ess") is not None else float("nan")
    print(f"{ns:7d} | {results_mc[-1]['mean']:12.8f} {results_mc[-1]['se']:10.3e} || "
          f"{results_is[-1]['mean']:12.8f} {results_is[-1]['se']:10.3e} {ess:10.1f}")

# calcul de l'amélioration en termes de nombre de simulations nécessaires
improvement_ratio = N_mc_converge / N_is_converge
improvement_pct = 100.0 * (1.0 - N_is_converge / N_mc_converge)

print("Tolérance (target se) =", se_target, "  |  Target diff =", diff_target)
print(f"Monte‑Carlo standard : convergence atteinte pour N = {N_mc_converge}")
print(f"Importance Sampling : convergence atteinte pour N = {N_is_converge}")
print(f"Amélioration (ratio N_MC / N_IS) = {improvement_ratio:.3f} "
          f"(réduction ≈ {improvement_pct:.1f}%)")

# --- Graphiques ---
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Convergence des moyennes
mc_means = [res["mean"] for res in results_mc]
is_means = [res["mean"] for res in results_is]
ax[0].plot([N*N_Factor**i for i in range(len(mc_means))], mc_means, label="Monte-Carlo standard", marker="o")
ax[0].plot([N*N_Factor**i for i in range(len(is_means))], is_means, label="Importance Sampling", marker="x")
ax[0].set_xscale("log")
ax[0].set_xlabel("Nombre de simulations (N)")
ax[0].set_ylabel("Prix estimé")
ax[0].set_title("Convergence des estimations")
ax[0].legend()
ax[0].grid()

# Erreur standard
mc_ses = [res["se"] for res in results_mc]
is_ses = [res["se"] for res in results_is]
ax[1].plot([N*N_Factor**i for i in range(len(mc_ses))], mc_ses, label="Monte-Carlo standard", marker="o")
ax[1].plot([N*N_Factor**i for i in range(len(is_ses))], is_ses, label="Importance Sampling", marker="x")
ax[1].set_xscale("log")
ax[1].set_xlabel("Nombre de simulations (N)")
ax[1].set_ylabel("Erreur standard")
ax[1].set_title("Erreur standard des estimations")
ax[1].legend()
ax[1].grid()

plt.tight_layout()
plt.show()
```
