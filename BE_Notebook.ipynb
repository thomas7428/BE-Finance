# Pricing des zéro-coupon dans un modèle à taux flottant : le modèle de Vasiček (Projet N° 1)

## Partie I : Le modèle et pricing des zéro-coupon

Soit le taux court $r=(r(t))_{t\ge 0}$ donné comme unique solution de l'équation différentielle stochastique suivante :
$$
dr(t)=b(t,r(t))\,dt+\sigma(t,r(t))\,dW_t,\qquad t\ge0,\quad r(0)=r_0,
$$
où $b,\sigma$ sont des processus adaptés, $W$ est un mouvement brownien défini sur $(\Omega,\mathcal{F},\mathbb{P})$ et $r_0>0$. On note $(\mathcal{F}_t)_{t\ge0}$ la filtration naturelle de $W$ (satisfaisant les conditions usuelles).

Le compte money-market associé au taux $r$ est donné par :
$$
B(t)=\exp\!\Big(\int_0^t r(s)\,ds\Big),\qquad t\ge0.
$$

En supposant que $\mathbb{P}$ est une mesure martingale équivalente (MME) pour la numéraire $B$, le prix à la date $t$ d'un zéro-coupon de maturité $T$ s'exprime par :
$$
p(t,T)=\mathbb{E}\!\left[\left. e^{-\int_t^T r(s)\,ds} \right|\mathcal{F}_t\right],\qquad t\in[0,T],\quad p(T,T)=1.
$$

Dans la suite, nous préciserons les hypothèses conduisant au modèle de Vasiček, résoudrons l'EDS du taux court et calculerons la formule fermée (ou semi-fermée) du prix $p(t,T)$.

**Question 1 :**

On fixe T>0 et on pose
$$
M_t:=p(t,T)B(t)^{-1}.
$$


a) Montrons que $M(t), t∈[0,T]$ est une martingale.

Par définition du prix du zéro‑coupon,
$$
p(t,T)=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
Donc car $B(t)=e^{\int_0^t r(s)\,ds}$ est $\mathcal{F}_t$-mesurable,
$$
M_t=\mathbb{E}\!\Big[\Big.e^{-\int_t^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big]\;e^{-\int_0^t r(s)\,ds}
=\mathbb{E}\!\Big[\Big.e^{-\int_0^T r(s)\,ds}\,\Big|\,\mathcal{F}_t\Big].
$$
En notant $M_T=e^{-\int_0^T r(s)\,ds}$, on a donc $M_t=\mathbb{E}[M_T\mid\mathcal{F}_t]$. Par la "Tower Property" de l'espérance conditionnelle : 
$$
\mathbb{E}[M_u\mid\mathcal{F}_t]=\mathbb{E}\big[\mathbb{E}[M_T\mid\mathcal{F}_u]\big|\mathcal{F}_t\big]=\mathbb{E}[M_T\mid\mathcal{F}_t]=M_t.
$$
Ainsi $M$ est une martingale (sous l'hypothèse d'intégrabilité suffisante de $M_T$).

b) On a que $M=(M_t)_{t\in[0,T]}$ est une martingale adaptée à la filtration naturelle de $W$ et $M_T\in L^2$, 



On a de plus le théorème de représentation des martingales :
___
Soit $B_t$ un mouvement brownien sur un espace de probabilité filtré standard $(\Omega,\mathcal{F},(\mathcal{G}_t)_{t\ge0},\mathbb{P})$, et soit $\mathcal{G}_t$ la filtration augmentée engendrée par $B$. Si $X\in L^2(\Omega,\mathcal{G}_\infty,\mathbb{P})$, alors il existe un processus prévisible $C$ adapté à $(\mathcal{G}_t)$ tel que
$$
X=\mathbb{E}[X]+\int_0^\infty C_s\,dB_s.
$$

Par conséquent,
$$
\mathbb{E}[X\mid\mathcal{G}_t]=\mathbb{E}[X]+\int_0^t C_s\,dB_s.
$$
___

En appliquant ce résultat à $X=M_T$ et en prenant $B=W$, il existe donc un processus prévisible $v=(v(t))_{t\in[0,T]}$ tel que
$$
M_t=M_0+\int_0^t v(s)\,dW_s,\qquad t\in[0,T],
$$
avec $\int_0^T v(s)^2\,ds<\infty$ p.s., et $v$ est unique à modifications négligeables.

**Question 2:**

Partant de $M_t=\dfrac{p(t,T)}{B(t)}$, on a $p(t,T)=M_tB(t)$. Par la formule d'Ito pour le produit de semimartingales :
$$
dp(t,T)=B(t)\,dM_t+M_t\,dB(t)+d\langle M,B\rangle_t.
$$
Or $B(t)=\exp\!\big(\int_0^t r(s)\,ds\big)$ est de variation finie, donc $dB(t)=B(t)r(t)\,dt$ ce qui nous donne aussi $\langle M,B\rangle_t=0$. 

De plus, d'après la représentation martingale $dM_t=v(t)\,dW_t$. On obtient donc :
$$
dp(t,T)=B(t)v(t)\,dW_t+M_tB(t)r(t)\,dt
= v(t)B(t)\,dW_t + p(t,T)r(t)\,dt,
$$
pour $t\in[0,T]$, ce qui donne bien l'expression souhaitée.

**Question 3 :**

Il n'y a pas d'opportunité d'arbitrage car $dr$ a été choisi directement de manière à avoir une MME.


**Question 4 :**

On suppose qu'il existe une fonction $F_T:[0,T]\times\mathbb{R}\to\mathbb{R}$ telle que $p(t,T)=F_T(t,r(t))$. On impose la régularité classique pour pouvoir appliquer la formule d'Itô et faire des égalités terme à terme :
- $F_T\in C^{1,2}([0,T)\times\mathbb{R})$ (continue en $t$, deux fois dérivable en $r$),
- croissances/conditions d'intégrabilité suffisantes (par exemple croissance au plus polynomiale) pour garantir l'application d'Itô et l'appartenance en $L^2$ de $M_T$.

IPP d'Itô :

$$
dM_t = e^{-\int_0^t r(s)\,ds}\cdot dF_T + F_T\cdot d(e^{-\int_0^t r(s)\,ds}) + <dF_T , d(e^{-\int_0^t r(s)\,ds})>
$$

On pose : $g(t) = \int_0^t r(s)\,ds$ d'où $g'(t) = r(t)$

Si on différence, on obtient : $d(e^{-g(t)})=-g'(t) \cdot e^{-g(t)} \cdot dt = -r(t) \cdot e^{-g(t)} \cdot dt $

De plus on a : $dg(t) = r(t) \cdot dt + 0 \cdot dW_t$
Nous n'avons donc pas de variations en $dW_t$ ce qui implique que nous n'avons pas de terme d'erreur quadratique.

Avec la formule d'Itô :
$$
dF_T = \partial_t F_T \cdot dt + \partial_r F_T \cdot dr + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 \cdot dt
$$
De plus : $d(e^{-g(t)}) = -r(t) \cdot e^{-g(t)} + 0 \cdot dW_t$ d'où $<...,...> = 0$

Ainsi : 
$$
dM_t = e^{-\int_0^t r(s)\,ds}\cdot (\partial_t F_T + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 - r(t) \cdot F_T)dt + e^{-\int_0^t r(s)\,ds} \cdot \partial_r F_T \cdot dr
$$
et $dr = b \cdot dt + \sigma \cdot dW_t$

D'où :
$$
dM_t = \phi_t \cdot dW_t <=> e^{-\int_0^t r(s)\,ds} \cdot (\partial_t F_T + 1/2 \cdot \partial_r^2 F_T \cdot \sigma^2 - r(t) \cdot F_T + \partial_r F_T \cdot b) = 0 \\
<=> \partial_t F_T + b \cdot \partial_r F_T + 1/2 \cdot \sigma^2 \cdot \partial_r^2 F_T + r \cdot F_T = 0
$$
De plus on a la condition terminale $F_T(T,r) = 1$ pour cette équation parabolique.

## Partie II : Le modèle de Vasiček

**Question 1 :**

On a : $dr(t) = (b + \beta \cdot r(t))dt + \sigma \cdot dW_t $
$$
Y_t := e^{-\beta t} r(t)
$$
Par Itô :
$$
dY_t = - \beta e^{-\beta t} r(t) \cdot dt + e^{-\beta t} dr(t) + 0 \\
dY_t = - \beta e^{-\beta t} r(t) \cdot dt + e^{-\beta t} b \cdot dt + e^{-\beta t} \beta r(t) \cdot dt + e^{-\beta t} \sigma dW_t \\
<=> Y_t = Y_0 + b \int_0^t e^{-\beta s}\,ds + \sigma \int_0^t e^{-\beta s}\,dW_s
$$
En inversant : $r_t = Y_r e^{\beta t}$
D'où :
$$
r_t = Y_0 e^{\beta t} + b e^{\beta t} \int_0^t e^{-\beta s}\,ds + \sigma e^{\beta t} \int_0^t e^{-\beta s}\,dW_s \\
r_t = \underbrace {r_0 e^{\beta t} + b e^{\beta t} \int_0^t e^{-\beta s}\,ds }_\textrm{Partie déterministe} + \underbrace {\sigma e^{\beta t} \int_0^t e^{-\beta s}\,dW_s}_\textrm{Partie gaussienne}
$$


**Question 2:**

On constate que $r_t$ suit une loi normale du fait de la partie gaussienne. La partie déterministe influence l'espérance.

```python
# Simulation et formules pour le modèle de Vasiček (question 5, Partie II)
# dr(t) = (b + beta * r(t)) dt + sigma dW_t, r(0)=r0
    
import numpy as np
import sys, subprocess
import matplotlib.pyplot as plt

# paramètres
r0 = 0.03
b = 0.02
beta = -0.3   # β
sigma = 0.01
T_maturity = 2.0   # maturité du zero-coupon (en années)
T_sim = 2.0        # horizon de simulation
n_steps = 400
n_paths = 30
dt = T_sim / n_steps
times = np.linspace(0, T_sim, n_steps + 1)

def mean_r(t, r0=r0, b=b, beta=beta):
    if beta == 0:
        return r0 + b * t
    return np.exp(beta * t) * r0 + (b / beta) * (np.exp(beta * t) - 1)

def var_r(t, sigma=sigma, beta=beta):
    if beta == 0:
        return sigma**2 * t
    return (sigma**2) * (np.exp(2 * beta * t) - 1) / (2 * beta)

# affichage de l'espérance et variance pour t fixes
print("Formules (pour t donné) :")
print("E[r(t)] =", "r0*e^{β t} + (b/β)(e^{β t}-1)  (si β != 0) ; sinon r0 + b t")
print("Var[r(t)] =", "σ^2 (e^{2β t}-1)/(2β)  (si β != 0) ; sinon σ^2 t")
print()
t_test = 1.0
print(f"E[r({t_test})] (num) = {mean_r(t_test):.6f}")
print(f"Var[r({t_test})] (num) = {var_r(t_test):.8f}")

# Les fonctions analytiques A,B et F_T sont laissées (vérification), mais la simulation
# et le calcul du prix suivent la discrétisation d'Euler-Maruyama et une discrétisation
# de l'intégrale pour p(t,T) (sans utiliser la forme explicite de F_T).

def B_func(t, T=T_maturity, beta=beta):
    tau = T - t
    if beta == 0:
        return tau
    return (1.0 - np.exp(-beta * tau)) / beta

def A_func(t, T=T_maturity, b=b, sigma=sigma, beta=beta):
    tau = T - t
    if beta == 0:
        return b * tau**2 / 2.0 - (sigma**2) * tau**3 / 6.0
    I1 = tau - (1.0 - np.exp(-beta * tau)) / beta
    I2 = tau - 2.0 * (1.0 - np.exp(-beta * tau)) / beta + (1.0 - np.exp(-2.0 * beta * tau)) / (2.0 * beta)
    return (b / beta) * I1 - 0.5 * (sigma**2) / (beta**2) * I2

def F_T(t, r, T=T_maturity, b=b, beta=beta, sigma=sigma):
    B = B_func(t, T=T, beta=beta)
    A = A_func(t, T=T, b=b, sigma=sigma, beta=beta)
    return np.exp(A - B * r)

# Test analytique (pour info)
print()
print("Prix analytique au temps 0 (p(0,T)) =", F_T(0, r0))

# 5) Simulation Euler-Maruyama pour trajectoires de r (utilisant bien dr = (b + beta r) dt + sigma dW)
rng = np.random.default_rng(12345)

def simulate_euler(n_paths=n_paths, n_steps=n_steps, dt=dt, r0=r0, b=b, beta=beta, sigma=sigma):
    rs = np.zeros((n_paths, n_steps + 1))
    rs[:, 0] = r0
    sqrt_dt = np.sqrt(dt)
    for i in range(n_steps):
        z = rng.normal(size=n_paths)
        # schéma d'Euler-Maruyama à partir de l'équation spécifiée
        rs[:, i+1] = rs[:, i] + (b + beta * rs[:, i]) * dt + sigma * sqrt_dt * z
    return rs

rs = simulate_euler()

# calculer prix zero-coupon le long des trajectoires par discrétisation de l'intégrale
# p(t,T) ≈ exp( - ∫_t^T r_s ds ) ; on utilise une somme à gauche cohérente avec Euler (Riemann gauche)
ps = np.empty_like(rs)
for i in range(n_steps + 1):
    # somme des r_j pour j = i..n_steps-1 (si i == n_steps -> somme vide = 0)
    integral = np.sum(rs[:, i:n_steps], axis=1) * dt
    ps[:, i] = np.exp(-integral)

# affichage : quelques trajectoires de r et p (calculées via discrétisation de l'intégrale)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
for k in range(min(10, n_paths)):
    plt.plot(times, rs[k], lw=1)
plt.title("Trajectoires r(t) (Euler-Maruyama)")
plt.xlabel("t"); plt.ylabel("r(t)")

plt.subplot(1, 2, 2)
for k in range(min(10, n_paths)):
    plt.plot(times, ps[k], lw=1)
plt.title(f"Prix zero-coupon p(t,{T_maturity}) via discrétisation de l'intégrale")
plt.xlabel("t"); plt.ylabel("p(t,T)")

plt.tight_layout()
plt.show()
```

```python
import math

# Estimation Monte-Carlo pour C_t = E[ e^{-∫_t^T r(s) ds} * (r_T - k)_+ | F_t ]
# Utilise les variables déjà définies dans le notebook (r0, rs, ps, times, n_steps, dt, sigma, b, beta, T_maturity, n_paths)


def cond_mean_rT(r_t, t, T=T_maturity, b=b, beta=beta):
    tau = T - t
    if beta == 0:
        return r_t + b * tau
    return math.exp(beta * tau) * r_t + (b / beta) * (math.exp(beta * tau) - 1)

def cond_var_rT(t, T=T_maturity, sigma=sigma, beta=beta):
    tau = T - t
    if beta == 0:
        return sigma**2 * tau
    return sigma**2 * (math.exp(2 * beta * tau) - 1) / (2 * beta)

def mc_price_from_state(r_t, t, k, n_sim=20000, use_IS=False, m_shift=0.5, seed=None):
    """
    Monte-Carlo price starting from state r(t)=r_t.
    Simule des prolongations de la trajectoire par Euler-Maruyama sur [t,T].
    Si use_IS True, les normales standards Z sont simulées ~ N(m_shift,1) (proposal)
    et on applique la pondération dP/dQ = exp(-m * sum Z + 0.5 * n * m^2).
    Renvoie (estimate, std_error, effective_sample_size_estimate).
    """
    rng_local = np.random.default_rng(seed)
    # indices et pas restants à simuler
    # nombre de pas restants = nombre de points discrets entre t et T
    # on déduit l'indice i_t correspondant à t dans `times`
    # si t n'est pas exactement sur un grid, on prend l'indice le plus proche vers la droite
    i_t = int(np.searchsorted(times, t, side='right') - 1)
    # si t est exactement times[i_t], steps_remaining = n_steps - i_t
    steps_remaining = n_steps - i_t
    sqrt_dt = math.sqrt(dt)
    payoffs = np.empty(n_sim)
    weights = np.ones(n_sim)
    for j in range(n_sim):
        r = r_t
        integral = 0.0
        sum_Z = 0.0
        # simuler increments Z_k ~ N(m,1) ou N(0,1)
        if use_IS:
            Zs = rng_local.normal(loc=m_shift, scale=1.0, size=steps_remaining)
            # pondération multiplicative = product p(z)/q(z) = exp(-m * sum Z + 0.5 * n * m^2)
            sum_Z = Zs.sum()
            weights[j] = math.exp(-m_shift * sum_Z + 0.5 * steps_remaining * (m_shift**2))
        else:
            Zs = rng_local.normal(size=steps_remaining)
        for z in Zs:
            dW = sqrt_dt * z
            r = r + (b + beta * r) * dt + sigma * dW
            integral += r * dt
        payoff = math.exp(-integral) * max(r - k, 0.0)
        payoffs[j] = payoff
    weighted = payoffs * weights
    estimate = weighted.mean()
    # standard error (using weighted sample variance)
    # var = (sum w*(x-mu)^2)/( (sum w) )  (not the unbiased form); std error approx = sqrt(var / n_eff)
    sum_w = weights.sum()
    mean_w = estimate
    var_w = (weights * (payoffs - mean_w)**2).sum() / sum_w
    # effective sample size approx
    ess = (weights.sum())**2 / (weights**2).sum() if use_IS else n_sim
    stderr = math.sqrt(var_w / ess)
    return estimate, stderr, ess

# 1) Prix au temps t général (formule) :
# C_t = E[ e^{-∫_t^T r(s) ds} * (r_T - k)_+ | F_t ]  -- on l'approxime numériquement ci-dessous.

# 2) Lien EDP : comme précédemment, si C_t = G(t, r(t)) alors G satisfait une PDE backward
#    ∂_t G + b(r,t) ∂_r G + 0.5 σ^2 ∂_r^2 G - r G = 0, terminale G(T,r) = (r-k)_+.
#    (même structure que pour le zéro-coupon, mais condition terminale non lisse).

# 3) Approximation MC : exemple numérique pour C_0 (t=0) en utilisant les trajectoires simulées
k = 0.06  # niveau de strike (à adapter)
# (a) Utiliser les chemins déjà simulés dans `rs` et `ps` : estimateur empirique pour C_0
C0_empirical = np.mean(ps[:, 0] * np.maximum(rs[:, -1] - k, 0.0))
print("MC (réutilisation chemins simulés) C_0 ≈", C0_empirical)

# (b) Simuler prolongations conditionnelles à partir de l'état actuel r0
est, se, ess = mc_price_from_state(r0, 0.0, k, n_sim=40000, use_IS=False, seed=2025)
print(f"MC (re-simulations de prolongations) C_0 ≈ {est:.6e}  ± {se:.2e}  (ESS≈{ess:.0f})")

# 4) Importance Sampling : situation utile :
#    Si k est élevé (payoff rare, p((r_T-k)>0) très petit), l'estimateur MC standard a grande variance.
#    On peut simuler sous une loi proposée qui favorise les trajectoires avec r_T grand
#    (ex. en décalant les normals des incréments par m_shift>0) puis corriger par le poids p/q.
k_large = 0.10
print("\nExemple importance sampling quand strike élevé (k = {:.2f})".format(k_large))
est_naive, se_naive, _ = mc_price_from_state(r0, 0.0, k_large, n_sim=40000, use_IS=False, seed=2026)
# choisir un m_shift modéré (heuristique) ; ici on prend m_shift=0.8 pour augmenter drift des incréments
est_is, se_is, ess_is = mc_price_from_state(r0, 0.0, k_large, n_sim=40000, use_IS=True, m_shift=0.8, seed=2027)
print(f"MC naive C_0 (k={k_large}) ≈ {est_naive:.6e} ± {se_naive:.2e}")
print(f"MC IS    C_0 (k={k_large}) ≈ {est_is:.6e} ± {se_is:.2e}  (ESS≈{ess_is:.0f})")
print("\nRemarque courte : choisir m_shift > 0 augmente la probabilité d'atteindre r_T > k. "
      "m_shift s'ajuste empiriquement (p.ex. partir de 0.5-1.0) ; on vérifie ESS et l'erreur "
      "pour s'assurer d'un gain en variance.")
```
